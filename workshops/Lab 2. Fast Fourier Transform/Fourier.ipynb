```python
T = len(data)/fs
```

```python
T
```

```python
t = np.linspace(0,T,len(data))
```

```python
t
```

```python
signal = data[:,0]
```

```python
plt.plot(t,data[:,0])
```

```python
A = []
f = []
for n in range(0,2500):
    a = 2/T * np.sum(signal * np.cos(2*np.pi*n*t))
    b = 2/T * np.sum(signal * np.sin(2*np.pi*n*t))
    A.append(np.sqrt(a**2+b**2))
    f.append(np.arctan(b/a))
```

```python
plt.plot(range(0,2500),np.log(A))
```

```python
plt.plot(range(0,2500),f)
```

```python
sig2 = np.zeros(len(signal)) + A[0]/2
for n in range(1,2500):
    sig2 += A[n]*np.cos(2*np.pi*n*t+f[n])
```

```python
plt.plot(sig2)
```

```python
from IPython.display import Audio
Audio(sig2, rate=fs)
```
