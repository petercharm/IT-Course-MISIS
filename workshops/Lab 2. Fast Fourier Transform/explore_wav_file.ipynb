```python
# Применяем к сигналу быстрое преобразование Фурье
res = fft(data,sample_rate)

# Из спектра получаем АЧХ - это абсолютные величины комплексных гармоник

res = [np.log(abs(i)) for i in res]

left = 0 
right = 1000

figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k') # указываем параметры фигуры для графика
plt.plot(range(left,right),res[left:right]) # График АЧХ (логорифмированный)

# В следующем цикле рисуются вертикальные линии и подписываются ноты
#for i,note in enumerate(notes): 
    #if left < note < right:
    #    plt.plot([note,note],[0,max(np.log(res[left:right]))])
    #    plt.text(note, 0, notes_labels[i%12])
        
Audio(data=data,rate=sample_rate)
```
