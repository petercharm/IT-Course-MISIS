```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Создаём сигнал
s = np.zeros(1000)

s[0:500] = 1

plt.plot(s)
```

```python
# Для сигнала нужна временная ось

t = np.linspace(0,1,1000)
```

```python
#s = np.sin(2*3.14*5.2*t)
```

```python
# Теперь на графике можно указать время в секундах

plt.plot(t,s)
```

```python
# Бустрое преобразование Фурье

A = []
fi = []

for k in range(0,500):
    a = 2/len(s)*np.sum(s * np.cos(2*3.14159*k*t))
    b = 2/len(s)*np.sum(s * np.sin(2*3.14159*k*t))
    A.append(np.sqrt(a**2 + b**2))
    fi.append(np.arctan(b/a))
```

```python
# Восстановление сигнала

s2 = np.zeros_like(s)

s2 = s2 + A[0]/2

for k in range(1,10):
    s2 = s2 + A[k]*np.cos(2*3.14159*k*t-fi[k])
```

```python
# Рисуем исходный сигнал и восстановленный

plt.plot(s)
plt.plot(s2)
```
